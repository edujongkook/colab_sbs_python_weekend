#### sin(x) / cos(x) 그리기

```python
import math

figure = plt.figure()
axes = figure.add_subplot(111)
list_x = []
list_sinx = []
list_cosx = []

for x in range(0, 100):
    list_x.append(x * 0.1)

for x in list_x:
    list_sinx.append(math.sin(x))
    list_cosx.append(math.cos(x))

axes.plot(list_x, list_sinx, linestyle='-')
axes.plot(list_x, list_cosx, 'r--')
plt.show()
```

### pie() 함수 (원형 그래프) -p344

```python
import matplotlib.pyplot as plt
figure = plt.figure()
axes = figure.add_subplot(111)
data = [1, 2, 3]
axes.pie(data)
plt.axis('equal')
plt.show()
```

```python
import matplotlib.pyplot as plt
figure = plt.figure()
axes = figure.add_subplot(111)
data = [1, 2, 3]
label = ['Good', 'Bad', 'Normal']
axes.pie(data, labels=label, autopct='%0d%%')
'''
autopct
%%  # %표시
%d  # 정수
%0.1f  # 소수점 1자리까지
%0.2f  # 소수점 2자리까지
%0.3f%%  # 소수점 3자리까지 + % 표시
'''
plt.legend(label, loc='center left')
plt.axis('equal')
plt.show()
```

#### ✅ 고양시 음식점 종류 원형 그래프 그리기  
csv파일 (goyang_food.csv)를 읽어서 음식점 종류의 분포를 원형 그래프로 표현하기

1. csv 파일 로드 (아래 링크에서 goyang_food.csv 다운로드후 코랩에 업로드)

공유 드라이브 링크 : https://drive.google.com/drive/folders/1t7-xAaEejn-0SvOetcLQvUNYxUhy4Lj4?usp=drive_link

```python
import csv
list_food = []  # 전체 음식점 정보가 담기는 리스트
with open('goyang_food.csv') as f:
    reader = csv.reader(f)
    for line in reader:  # 한줄씩 csv 파일을 읽어서
        list_food.append(line)  # 리스트에 추가

food_pie = list_food[1:]  # 맨 처음 요소에는 제목이 들어 있어서 슬라이스로 제외시킴
foods = []  # 음식점 종류만 담을 리스트
for food in food_pie:  # 음식점 전체 정보 리스트를 순회하면서
    foods.append(food[4])  # 음식점 종류가 담긴 5번째 요소만 추가
```

#### ✅ 롤 챔프 랭커 wordcloud  
게임 롤의 랭커들의 아이디와 점수가 들어있는 ranker.csv 파일을 불러와 각 챔피온의 이름과 점수로 wordcloud 그리기

load_csv_file 함수

```python
import csv

def load_csv_file(file_name):
    """ 이 함수는 csv파일 이름을 받아서
    한줄씩 리스트에 넣어서 돌려준다.
    ex) load_csv_file('file_name.csv') -> csv_list
    이 리스트는 요소로 리스트가 들어있다
    """
    csv_list = []
    with open(file_name) as f:
        reader = csv.reader(f)
        for line in reader:
            csv_list.append(line)

    return csv_list
```
