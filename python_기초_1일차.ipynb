## 주석 (comment)

✅ p29 예제

```python
import math

def get_area(radius):
    """
    개요: 반지름을 전달하면 원의 넓이를 반환하는 get_area() 함수
    작성자 : 김종국
    최초작성일 : 2023-06-17
    """
    area = math.pi * math.pow(radius, 2)
    return area
```

```python
radius = 1.5
# get_area() 함수 호출 결과를 area 변수에 저장
area = get_area(radius)
print(area)
print(get_area.__doc__) #Docstring 확인
```

### mutable / immutable -p47

- mutable 은 수정해도 저장된 위치가 바뀌지 않음 (list / set / dict)

```python
ex_list = [1, 2, 3 ,4 ,5]
id(ex_list)
```

```python
ex_list.append(6)
id(ex_list)
```

- immutable 은 값에 손을 대는 순간
저장되는 위치가 새로운 곳으로 바뀜 (int / float / str / tuple)

```python
s = "hello"
id(s)
```

```python
s = "world"
id(s)
```

- mutable 변수를 대입하는 경우 같은 위치를 가르키는 두개의 변수가 됨
따라서 수정사항도 함께 공유

```python
ex_list_new = ex_list
```

```python
id(ex_list)
```

```python
id(ex_list_new)
```

#### ⚡ 새로운 저장공간에 복사하고 싶은 경우 copy 모듈 사용

```python
import copy
ex_list_new = copy.deepcopy(ex_list)
# 새로운 공간을 만들어 값을 복사하는 것을 deepcopy (깊은 복사)
# 저장되있는 장소의 주소만 복사하는것을 Shallow copy(얕은 복사) 라고 함
```

```python
id(ex_list)
```

```python
id(ex_list_new)
```

### 문자열 포맷팅 (formatting)

#### % 연산자 - p56

```python
print('%d' % 10)
print('%o' % 10)
print('%x' % 10)
print('%f' % 3.14)
print('%s' % 'Python')
print('%d %d' %(10,20))
```

#### format() 매소드 - p59

```python
my_name = 'My name is'
print('{} {}'.format(my_name, 'James'))
print('{0} {2}'.format(my_name, 'James', 'John'))
print('{} {name}'.format(my_name, name='James'))
print('{} {}. I\'m {} years old.'.format(my_name, 'James', 25))
print('{} {name}. I\'m {age} years old.'.format('my name is', name='James', age=25))
```

#### f-strings -p61

```python
who = 'you'
how = 'haapy'
print(f'{who} make me {how}')
age = 25
print(f'내년엔 {age+1}살이 됩니다')
```

```python
print(f'{"left":<10}')
print(f'{"center":^10}')
print(f'{"right":>10}')
```

```python
pi = 3.141592653589
print(f'{pi:10.10f}')
print(f'{pi:10.4f}')
print(f'{pi:10.2f}')
print(f'{pi:.4f}')
```
